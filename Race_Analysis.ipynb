## DATA CLEANING

### (i) Only want USA Race
### (ii) Event Dis :-  50k or 50mi
### (iii) Year :- 2020

```python
df[df['Event distance/length']=='50mi']
```

```python
# combine 50K/50mi with isin

df[df['Event distance/length'].isin(['50km','50mi'])]
```

```python
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event']==2020)]
```

```python
df[df['Event name']=='Everglades 50 Mile Ultra Run (USA)']
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)=='USA']
```

```python
# All requirements together and name it df2

df2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event']==2020) & 
         (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)=='USA')]
```

```python
df2.head(4)
```

```python
df2.shape
```

```python
# remove (USA) from event name

df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
df2.head(4)
```

```python
# Getting athlete age 

df2['athlete_age'] = 2020 - df2['Athlete year of birth']
df2.head(3)
```

```python
# removing h from athlete performance

df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
df2.head(3)
```

#### Drop columns : Athlete club, Athlete country, Athlete year of birth, Athlete Age category

```python
df2 = df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'],axis=1)
```

```python
df2.head(3)
```

#### Cleaning null values

```python
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

#### Checking Duplicates

```python
df2[df2.duplicated()==True]
```

#### Reset Index

```python
df2.reset_index(drop=True)
```

#### Checking Data types and fixing them

```python
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
# renaming columns

df2 = df2.rename(columns = {'Year of event': 'year',
                        'Event dates': 'race_day',
                        'Event name': 'race_name',
                        'Event distance/length': 'race_length',
                        'Event number of finishers': 'race_number_of_finishers',
                        'Athlete performance': 'athlete_performance',
                        'Athlete gender': 'athlete_gender',
                        'Athlete average speed': 'athlete_average_speed',
                        'Athlete ID': 'athlete_id' 
})
```

```python
df2.head(4)
```

```python
#reorder columns

df3 = df2[['race_day','race_name','race_length','race_number_of_finishers','athlete_id','athlete_gender','athlete_age',
           'athlete_performance','athlete_average_speed']]
```

#### Difference in speed for the 50k, 50mi male to female

```python
df3.groupby(['race_length','athlete_gender'])['athlete_average_speed'].mean()
```

#### What age groups are the best in the 50m Race (20+ races min)

```python
df3.query('race_length=="50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False).query('count>19').head(10)
```

#### What age group are the worst in the 50m Race (10+ races min) (show 20)

```python
df3.query('race_length=="50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=True).query('count>9').head(15)
```
